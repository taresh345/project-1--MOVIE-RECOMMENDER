```python
# to convert str to list we import a new library called ast 
import ast
def convert(obj):
    l=[]
    a=ast.literal_eval(obj)
    for i in a: 
        l.append(i['name'])
    return l 
```

```python
df['genres']=df['genres'].apply(convert)
df['keywords']=df['keywords'].apply(convert)

```

```python
df.iloc[0].cast
#  series of dictionary with each dictionary containg character info !
#
```

```python
def convert_crew_cast(obj):
    l=[]
    counter= 0
    b=ast.literal_eval(obj)
       
    for i in b:
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
df['cast']=df['cast'].apply(convert_crew_cast)
```

```python
def fetch_director(obj):
    l=[]
    crew=ast.literal_eval(obj)
       
    for i in crew:
        if i['job']=='Director':
            
            l.append(i['name'])
            break
    return l
# only those names who are director
# 
```

```python
df['crew']=df['crew'].apply(fetch_director)
```

```python
import nltk 
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text ):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#  repeeat vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')

vectors=cv.fit_transform(new_df['tags']).toarray()
cv.get_feature_names() # len 5000
# (4806,5000)
```

```python
# calculate cpsine distance between each vector for each movie
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
# movie 1   -avatar    then similiarity 0th row
# sort the row   in ascending order
# grab the first 5  similarity indexes
# ________________________________ 1).  grab the index of the movie.       2) .find distances =similarity[index] 3.) sort the distances 
similarity[0]
```

```python
# enumerate?  enumerate(iterable, start=0) The enumerate object yields pairs containing a count (from start, which
# defaults to zero) and a value yielded by the iterable argument. 
# enumerate is useful for obtaining an indexed list:
#     (0, seq[0]), (1, seq[1]), (2, seq[2]), ...
#  we are loosing index postion which are aligned with original title of the movies   so we use enumerate and sort by the value 
```

```python
sorted(list(enumerate(similarity[1216])),reverse=True,key = lambda x: x[1])

#  dummy code
# index=movie index


# closeness of the input movie to everymovie  vector   sorted  by value  while not changing the index !
```

```python
def recommend(movie):
    index = new_df[new_df['original_title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].original_title)   #find movie name with indexes
```

```python
recommend('Standard Operating Procedure')
```
